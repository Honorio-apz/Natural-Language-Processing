# **Experimentación con Algoritmo Jaro Winkler**

```python
# Python3 implementation of above approach 
from math import floor, ceil 
  
# Function to calculate the 
# Jaro Similarity of two s 
def jaro_distance(s1, s2): 
      
    # If the s are equal 
    if (s1 == s2): 
        return 1.0
  
    # Length of two s 
    len1 = len(s1) 
    len2 = len(s2) 
  
    # Maximum distance upto which matching 
    # is allowed 
    max_dist = floor(max(len1, len2) / 2) - 1
  
    # Count of matches 
    match = 0
  
    # Hash for matches 
    hash_s1 = [0] * len(s1) 
    hash_s2 = [0] * len(s2) 
  
    # Traverse through the first 
    for i in range(len1): 
  
        # Check if there is any matches 
        for j in range(max(0, i - max_dist),  
                       min(len2, i + max_dist + 1)): 
              
            # If there is a match 
            if (s1[i] == s2[j] and hash_s2[j] == 0): 
                hash_s1[i] = 1
                hash_s2[j] = 1
                match += 1
                break
  
    # If there is no match 
    if (match == 0): 
        return 0.0
  
    # Number of transpositions 
    t = 0
    point = 0
  
    # Count number of occurances 
    # where two characters match but 
    # there is a third matched character 
    # in between the indices 
    for i in range(len1): 
        if (hash_s1[i]): 
  
            # Find the next matched character 
            # in second 
            while (hash_s2[point] == 0): 
                point += 1
  
            if (s1[i] != s2[point]): 
                point += 1
                t += 1
    t = t//2
  
    # Return the Jaro Similarity 
    return (match/ len1 + match / len2 + 
            (match - t + 1) / match)/ 3.0
  
# Driver code 
#s1 = ""
#s2 = ""
  
# Prjaro Similarity of two s 
#print(round(jaro_distance(s1, s2),6)) 
  
# This code is contributed by mohit kumar 29 
```

```python
df_tesis_titulo = Tesis[['Titulo']]
df_tesis_abstract=Tesis[['abstract']]

df_paper_Area=Papers[['Area']]
df_paper_Titulo=Papers[['Titulo']]
df_paper_Keywords=Papers[['Keywords']]
df_paper_ABSTRACT=Papers[['ABSTRACT']]
```

# Algoritmo Jaro Winkler

```python
cadena=[]
for tesis in range(df_tesis_titulo.shape[0]):
  str1, str2="",""
  ts_titulo=str1.join(df_tesis_titulo.iloc[tesis])
  ts_abstract=str2.join(df_tesis_abstract.iloc[tesis])
  txt_research=ts_titulo+" "+ts_abstract
  #print(txt_research)
  #print(ts_abstract)
  for paper in range(df_paper_Area.shape[0]):
    str1,str2,str3,str4="","","",""
    pp_area=str1.join(df_paper_Area.iloc[paper])
    pp_titulo=str2.join(df_paper_Titulo.iloc[paper])
    pp_keywords=str3.join(df_paper_Keywords.iloc[paper])
    pp_abstract=str4.join(df_paper_ABSTRACT.iloc[paper])
    #print(pp_abstract)
    research_area=round(jaro_distance(s1=txt_research, s2=pp_area),6)
    research_titulo=round(jaro_distance(s1=txt_research, s2=pp_titulo),6)
    research_keywords=round(jaro_distance(s1=txt_research, s2=pp_keywords),6)
    research_abstract=round(jaro_distance(s1=txt_research, s2=pp_abstract),6)

    cadena.append(tesis)
    cadena.append(paper)
    cadena.append(research_area)
    cadena.append(research_titulo)
    cadena.append(research_keywords)
    cadena.append(research_abstract)
```

# Exportar resultados

```python
import csv
with open('JaroW_Carmen.csv', 'w') as csvfile:
    fieldnames = ['Id_tesis','Id_paper','paper_area','paper_titulo','paper_keywords','paper_abstract']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    word=0
    for i in range(int(len(cadena)/6)):
        Id_tesis=cadena[word]
        Id_paper=cadena[word+1]
        research_area=cadena[word+2]
        research_titulo=cadena[word+3]
        research_keywords=cadena[word+4]
        research_abstract=cadena[word+5]
     
        writer.writerow({'Id_tesis': Id_tesis,
                         'Id_paper':Id_paper,
                         'paper_area':research_area,
                         'paper_titulo':research_titulo,
                         'paper_keywords':research_keywords,
                         'paper_abstract':research_abstract})
        cont=word+5
        word=cont+1
```
